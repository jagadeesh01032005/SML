```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("/content/breast_cancer_survival.csv")
print(data)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for col in data.columns:
    if data[col].dtype == 'object':
        le = LabelEncoder()
        data[col] = le.fit_transform(data[col])
```

```python
data
```

```python
x=data.drop('Patient_Status',axis=1)
print(x)
y=data['Patient_Status']
```

```python
x
y
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Train SVM, KNN, and Logistic Regression and calculate accuracy
def train_and_evaluate(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    return accuracy
```

```python
# Define the models
models = {
    "SVM": SVC(),
    "KNN": KNeighborsClassifier(),
    "Logistic Regression": LogisticRegression()
}
```

```python
# Train and evaluate each model before PCA
print("Accuracy before PCA:")
for name, model in models.items():
    accuracy = train_and_evaluate(model, X_train, X_test, y_train, y_test)
    print(f"{name}: {accuracy:.2f}")
```

```python
# Reduce dimensions with PCA
pca = PCA(n_components=0.95)  # retain 95% of variance
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
# Train and evaluate each model after PCA
print("\nAccuracy after PCA:")
for name, model in models.items():
    accuracy = train_and_evaluate(model, X_train_pca, X_test_pca, y_train, y_test)
    print(f"{name}: {accuracy:.2f}")
```
