```python
import pandas as pd
d = pd.read_csv("/content/diabetes_data_upload.csv")
print(d)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
target_variable = 'class'
for column in d.columns:
    if column != target_variable:
        plt.figure()
        sns.histplot(data=d, x=column, hue=target_variable, kde=True)
        plt.title(f'Distribution of {column} by {target_variable}')
        plt.xlabel(column)
        plt.ylabel('Count')
        plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
d_normalized = pd.DataFrame(scaler.fit_transform(d.select_dtypes(include=['number'])), columns=d.select_dtypes(include=['number']).columns)
d_normalized = d_normalized.join(d.select_dtypes(exclude=['number']))
print(d_normalized)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
string_columns = d.select_dtypes(include=['object']).columns
for column in string_columns:
    d[column + '_encoded'] = le.fit_transform(d[column])
print(d.head())
```
