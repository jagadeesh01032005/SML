```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
data = pd.read_csv("/content/housing.csv")
print(data.head(2))

container = ['latitude' , 'total_rooms' , 'housing_median_age' , 'total_bedrooms']
result = data[container]

mle_params = {}


for i in container:
    mean = np.mean(data[i])
    variance = np.var(data[i])
    mle_params = {'mean' : mean , 'variance' : variance}
```

```python
bins = [5,10,15,20]
for i in container:
    plt.hist(data[i],bins=bins,alpha=0.7,color="red",density="True")
    plt.xlabel(i)
    plt.ylabel('Density')
    plt.show()
```
